```python
#import Essential Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
df=sns.load_dataset('tips')
df.head()
```

```python
df.shape
```

```python
#scatter plot:- Scatter plot is used to show relationship b/w two variables.
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.title("SCATTER PLOT")
plt.show()
```

```python
#Line plot:- Line plot is used to show trend visualization
sns.lineplot(x="size",y="total_bill",data=df)
plt.title("LINE PLOT")
plt.show()
```

```python
#Bae plot:- Bar plot is used to show avarge value.
sns.barplot(x="day",y="total_bill",data=df)
plt.title("BAR PLOT")
plt.show()
```

```python
#Count plot:- count plot count the frequencies of categories.
sns.countplot(x="day",data=df)
plt.title("COUNT PLOT")
plt.show()
```

```python
#Histogram plot:- Histogram plot is used to show distribution of data.
sns.histplot(df["total_bill"],bins=100)
plt.title("HISTOGRAM PLOT")
plt.show()
```
