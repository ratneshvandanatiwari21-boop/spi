```python
average_score_by_gender=df.groupby('Gender')['Final_Score'].mean()
print(average_score_by_gender)
```

```python
average_score_by_parent_education=df.groupby('Parental_Education')['Final_Score'].mean()
print(average_score_by_parent_education)
```

```python
average_score_by_internet_access=df.groupby('Internet_Access')['Final_Score'].mean()
print(average_score_by_internet_access)
```

```python
total_study_hours=df.groupby('Gender')['Study_Hours'].sum()
print(total_study_hours)
Average_study_hours=df.groupby('Gender')['Study_Hours'].mean()
print(Average_study_hours)
```

```python
percent_above_50=((df['Final_Score']>50).sum())*100/df['Student_ID'].nunique()
print('Pass Percent=',percent_above_50)
```

```python
percent_above_75=((df['Final_Score']>75).sum())*100/df['Student_ID'].nunique()
print('Pass Percent=',percent_above_75)
```

```python
avg=df[df['Final_Score']>=90]['Final_Score'].mean()
print('Average Top Performance Score',avg)
```
