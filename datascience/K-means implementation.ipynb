```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data={
    'Age':[22,25,47,52,46,56,23,27,48,55],
    'income':[15000,16000,25000,27000,24000,26000,18000,19000,28000,29000]
}
df=pd.DataFrame(data)
df.head(2)
```

```python
#aply k-mean algo for given dataset
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2)
df['Cluster']=kmeans.fit_predict(df)
```

```python
df.head()
```

```python
#visualize the data
plt.scatter(df['Age'],df['income'],c=df['Cluster'],cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('K-means Cluster')
plt.show()
```
