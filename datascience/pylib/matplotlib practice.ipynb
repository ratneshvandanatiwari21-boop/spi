```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
# Example application-1
x=[1,2,3,4]
y=[4,5,1,5]
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("My Graph")
plt.show()
```

```python
#Example application-2
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'k', label="Line 1", linewidth=5)
plt.plot(x2,y2,'y', label="Line 2", linewidth=5)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Epic Info")
plt.grid(color='b')
plt.legend()
plt.show()
```

```python
#Example application-3
st=['Amit','Rahul','Pooja','ratnesh']
num=[33,45,0,97]
plt.bar(st,num)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Mark Info")
plt.show()
```

```python
#Area plot aaplication Example
months=['Jan','Feb','March','Aprail','May','june']
sales=[1200,400,1000,400,3000,2000]
plt.fill_between(months,sales,color="orange")
plt.xlabel("Sales")
plt.ylabel("Months")
plt.title("Sales Growth")
plt.grid(color="pink")
plt.show()
```
